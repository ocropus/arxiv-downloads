```python
import webdataset as wds
```

```python
ds = wds.WebDataset("https://storage.googleapis.com/ocro-arxiv/pages/arxiv-pages-{000001..000100}.tar").decode("rgb")
```

```python
figsize(20, 20)
for i, sample in enumerate(ds.slice(0, 4)):
    subplot(2, 2, i+1)
    # title(repr(sample["mp"][0]))
    imshow(sample["jpg"])
```

```python
gt = sample["mp"]
print(gt)
```

```python
imshow(sample["jpg"])
```

```python
for i, line in enumerate(gt):
    text = [word["text"] for word in line]
    print(f"{i:6d}:", " ".join(text))
```
